```python
import pandas as pd
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

# Я выбрал экспоненциальное распределение c параметром lambda = 3

```python
# Описание функции
expon_rv = sts.expon(loc=0,scale=1/3)

#генерируем выборку объема 1000
sample = expon_rv.rvs(size=1000)
```

```python
fig, ax = plt.subplots(1,1)

#определяем массив точек для построения плотности распределения
x = np.linspace(expon_rv.ppf(0.01), expon_rv.ppf(0.99),200)

#строим гистрограмму выборки и теоретическую функцию распределения
ax.hist(sample,color='m',density=True,alpha=0.8,histtype='stepfilled')
ax.plot(x,expon_rv.pdf(x),'g-',lw=2,alpha = 0.8,label='theo. prob. density function')
plt.xlabel('x')
plt.ylabel('Prob. Density function')
ax.legend(loc='best')
plt.show()
```

# Распределение выборочного среднего случайной величины при разных объёмах выборок

Мы напрминаем что для экспоненциального распределения с параметром lambda (X ~ exp(lambda), 
мат. ожидание $$MX = \frac{1}{\lambda}$$, и 
дисперсия     $$DX = \frac{1}{ \lambda^2}$$

 Для выборки объёма n, параметры нормального распределения приближающего выборку имеют вид:
 мат. ожидание $$MX = \frac{1}{\lambda}$$
 дисперсия     $$DX = \frac{1}{n * \lambda^2}$$
 То есть среднее квад. отклонение $$\sigma = \frac{1}{\lambda * \sqrt n}$$.
 
 Мы получили аппроксимацию $$N(\frac{1}{\lambda},\frac{1}{\lambda * \sqrt n})$$

## При  n = 5

```python
n=5
means = np.array([])#обозначим массив выборочных средних
i=1
while i <= 1000:#цикл генерации случайных выборок количества n из функции
    sample = expon_rv.rvs(n) #генерация выборки
    newMean = sum(sample)/n #определение выборочного среднего (в.с.)
    means = np.append(means, newMean) #добавляем в массив новое значение в.с.
    i+=1
    
#определим параметры нормального распределения описывающие массив 1000 в.с. при n=5
MX = 0.33333 # MX = 1/lambda           lambda = 3
DX = 0.02222 # DX = 1/(n*lambda^2)     n = 5
sigma = math.sqrt(DX)  #сигма для нормального распределения выборки средних
norm_rv = sts.norm(MX, sigma) #определяем нормальное распределение с расчетными характеристиками
pdf = norm_rv.pdf(x) #получаем плотность распределения


#строим гистограмму массива средних при n=5 и плотность описывающую этот массив нормального распределения
plt.plot(x, pdf, label='PDF(n=5)')
plt.hist(means,density=True,alpha=0.8,histtype='stepfilled')
plt.ylabel('Prob. density function')
plt.xlabel('x')
plt.legend(loc='best')
plt.show()
```

```python
femaleNames = data[data["Sex"] == 'female']["Name"]


femaleNames = list(femaleNames)

femaleNames = list(map(lambda x:x.split(","),femaleNames))
```

```python
femaleNames = np.array(femaleNames)
femaleNames
```

```python
fullname = []
for name in femaleNames:
    fullname.append(name[1].strip())
#fullname

firstname = []
for name in fullname:
    name = name.split()
    if name[0] == 'Miss.':
        for elt in name[1:]: 
            if elt[0] != '"' and elt[0] != '(':
                firstname.append(elt)
            if elt[0] == '(':
                break
    else :
        for elt in name[1:]: 
            if elt[0] == '(':
                nnew = ''.join(elt[1:].split(')'))
                firstname.append(nnew)
                break
#firstname
    
```

```python
popularName = stats.mode(firstname)

with open('6.txt','w', encoding = 'utf-8') as infile:
    print(*popularName[0], file = infile, end='')
```

# Decision trees

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("titanic.csv",index_col='PassengerId', header = 0)
```

```python
data[:10]
```

```python
#ATTRIBUTES
clean_data = data.dropna(subset = ['Age'])
X = clean_data[['Pclass','Fare','Age','Sex']].values

#lABEL
y = clean_data[['Survived']]
y = np.squeeze(np.asarray(y))
```

```python
features = ['Pclass','Fare','Age','Sex']
```

```python
#Sex conversion into binar objects
sex_mapping = {
    'male':0,
    'female':1
}
for i in range(len(X)):
    X[i,3] = sex_mapping[X[i,3]]

#Split the data into train and test 
#X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2)
```

```python
# We create a model
clf = DecisionTreeClassifier(random_state=241)
#We train our model
clf.fit(X,y)
```

```python
#predictions
      #predictions = clf.predict(X_test)
      #accuracy = accuracy_score(y_test,predictions)
```

```python
#Features importance

importances = clf.feature_importances_

feat_importance = []
for i, value in enumerate(importances):
    feat_importance.append((i,value))
    
# We sort the features indexes from the most important to the least important     
feat_importance.sort(key = lambda x: x[1], reverse = True)
print(feat_importance)
```

```python
# We return the first two most important features

with open('answer.txt', 'w', encoding='utf-8') as fileout:
    print(
        features[feat_importance[0][0]],
        features[feat_importance[1][0]],
        file = fileout,
        end=''
    )
```
